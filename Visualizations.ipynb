```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
# Functional method to matplotlib
plt.plot(x,y)
#plt.show()
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("title")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object Oriented Method of matplotlib
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel("x axes label")
axes.set_ylabel("y axes label")
axes.set_title("title")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,1.0,1.0])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title("Large Plot")
axes2.plot(y,x)
axes2.set_title("Small plot")
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
#axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=3,ncols=3)
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title("first")
axes[1].plot(y,x)
axes[1].set_title("second")
```

```python
fig = plt.figure(figsize=(6,2),dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(6,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
fig.savefig('foo.png',dpi=200)
```

```python
fig = plt.figure(dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label="X squared")
ax.plot(x,x**3, label="X cubed")
ax.set_title("Title")
ax.set_xlabel("xlabel")
ax.set_ylabel("ylabel")
# ax.legend(loc=0)
ax.legend(loc=(0.05,0.5))
```
